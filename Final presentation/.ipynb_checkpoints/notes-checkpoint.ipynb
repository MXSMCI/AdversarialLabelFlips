Imma try this: https://hackernoon.com/guide-to-handling-internet-archives-cdx-server-api-response-c469df5b81f4 to get a list from the cdx api

```python
class Snapshot(dict):
   def __init__(self, urlkey=None, timestamp=None, original=None, mimetype=None, statuscode=None, digest=None, length=None):
       super(Snapshot, self).__init__()
       self['urlkey'] = urlkey
       self['timestamp'] = timestamp
       self['original'] = original
       self['mimetype'] = mimetype
       self['statuscode'] = statuscode
       self['digest'] = digest
       self['length'] = length
       self['snapshot_url'] = 'http://web.archive.org/web/%s/%s/' % (timestamp, original)
```

Now let's read in th the cdx text response:

```python
import requests

def unique_snapshots(url):
    # calls the webb archive cdx api for a specific url
    # returns a dictionary of all unique versions of that site with start and end dates.
    
    res = requests.get("https://web.archive.org/cdx/search/cdx?url=" + url)

    #split the text into a list of raw_snapshots
    raw_snapshots = res.text.split('\n')

    #init list
    snapshots = []

    #split each snapshot into a list of different data fiedls
    for snapshot in raw_snapshots:
        snapshot_items = snapshot.split(' ')
        if len(snapshot_items) == 7:
            snap = Snapshot(snapshot_items[0], snapshot_items[1], snapshot_items[2], snapshot_items[3], snapshot_items[4], snapshot_items[5], snapshot_items[6])
        #else:
        #    print("len(snapshot_items) not equal 7!?", snapshot_items)
        snapshots.append(snap)

    #turn "timestamp" into two fields: "since" and "until"
    snapshots_since_until = []
    for s in snapshots:
        time = int(s.get("timestamp"))
        s.update({"since": time, "until" : time})
        snapshots_since_until.append(s)
    
    # remove duplicate items in the list. only keep the first and last of a kind.
    # do it by putting it in a dict (ie hashmap).
    # use the digest provided by cdx as a key!

    unique = {}   
    for s in snapshots:
        # if s is already in unique
        if s["digest"] in unique:
            duplicate = unique.pop(s["digest"])

            #print(duplicate["since"],s["since"])
            #print(min(duplicate["since"],s["since"]))

            new_since = min(duplicate["since"],s["since"]) # update since date to the smaller of both
            new_until = max(duplicate["until"],s["until"]) # update until date to the larger of both

            s.update({"since": new_since, "until": new_until})
            unique.update({s["digest"] : s})
        else:    
            unique.update({s['digest']: s})


    return unique
```

```python
desktop_price_pics = unique_snapshots("www.zerodium.com/images/zerodium_prices.png")
mobile_price_pics = unique_snapshots("www.zerodium.com/images/zerodium_prices_mobiles.png")
```

```python
from PIL import Image
from io import StringIO
def add_imgs(snapshot_dict):
    # download the pictures and add them to the snapshots
    for digestkey in snapshot_dict:
        snapshot = snapshot_dict.get(digestkey)
        url = snapshot['snapshot_url']
        img = Image.open(requests.get(url, stream = True).raw)
        snapshot.update({"image":img})

add_imgs(mobile_price_pics)
add_imgs(desktop_price_pics)
```

so, from comparing pictures it seems that the categories in the "periodic table" change quite significantly over time.
Some dissapear while others appear. But more importantly they sometimes change slightly.

Maybe a good way to think about this, is that the *bins* in the "exploit space" are changing. It probably makes most sense to look at the changes of prices by looking at a specific 0-day that was found, and seeing how its price changed.
Because the price of an exploit can change in one of two ways:
1. the price of the bin changes.
2. the size of the bins change, i.e. the exploit now covered by another bin.

Also interesting: maybe they know of the 0-day before google zero does, resulting in a decrease in price shortly before the public 0-day discovery. -> Confirmed: (https://twitter.com/Zerodium/status/1260541578747064326) they obviously see supply increase, simply be seeing the offers flooding in.
